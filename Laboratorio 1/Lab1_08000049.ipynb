## Problema 1
Desarrolle una funcion la cual reciba dos parametros, una imagen y un entero llamado color, la
funcion debe devolver una imagen la cual tenga activos los canales de color segun los siguientes
puntos:
1. Si el parametro color vale 1, la imagen debe mostrar activos unicamente el color azul.
2. Si el parametro color vale 2, la imagen debe mostrar activos unicamente el color verde.
3. Si el parametro color vale 3, la imagen debe mostrar activos unicamente el color rojo.
4. Si el parametro color vale 10, la imagen debe mostrar activos unicamente los colores rojo y
verde.
5. Si el parametro color vale 20, la imagen debe mostrar activos unicamente los colores verde y
azul.
6. Si el parametro color vale 30, la imagen debe mostrar activos unicamente los colores azul y
rojo.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def filtrar_colores(imagen, color):
   
    imagen_nueva1 = np.zeros_like(imagen)
    
    if color == 1:
        # Azul
        imagen_nueva1[:, :, 0] = imagen[:, :, 0]
    elif color == 2:
        # verde
        imagen_nueva1[:, :, 1] = imagen[:, :, 1]
    elif color == 3:
        # rojo
        imagen_nueva1[:, :, 2] = imagen[:, :, 2]
    elif color == 10:
        # rojo y verde
        imagen_nueva1[:, :, 1] = imagen[:, :, 1]
        imagen_nueva1[:, :, 2] = imagen[:, :, 2]
    elif color == 20:
        # verde y azul
        imagen_nueva1[:, :, 0] = imagen[:, :, 0]
        imagen_nueva1[:, :, 1] = imagen[:, :, 1]
    elif color == 30:
        # azul y rojo
        imagen_nueva1[:, :, 0] = imagen[:, :, 0]
        imagen_nueva1[:, :, 2] = imagen[:, :, 2]
    else:
        
        return imagen
    
    return imagen_nueva1
```

## Problema 2

En el .zip del laboratorio se le compartio un conjunto de imagenes en escala de grises (imagen1, imagen2, perro) estas imagenes fueron creadas utilizando una escala de grises en 3D, cree una funcion que dadas las 3 imagenes se construya la imagen original a color.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def reconstruir_imagen_color(imagen1_paths, imagen2_paths, perro_paths):
    # Cargar imagenes
    imagen1_azul = cv2.imread(imagen1_paths[0], cv2.IMREAD_GRAYSCALE)
    imagen1_verde = cv2.imread(imagen1_paths[1], cv2.IMREAD_GRAYSCALE)
    imagen1_rojo = cv2.imread(imagen1_paths[2], cv2.IMREAD_GRAYSCALE)
    
    imagen2_azul = cv2.imread(imagen2_paths[0], cv2.IMREAD_GRAYSCALE)
    imagen2_verde = cv2.imread(imagen2_paths[1], cv2.IMREAD_GRAYSCALE)
    imagen2_rojo = cv2.imread(imagen2_paths[2], cv2.IMREAD_GRAYSCALE)
    
    perro_azul = cv2.imread(perro_paths[0], cv2.IMREAD_GRAYSCALE)
    perro_verde = cv2.imread(perro_paths[1], cv2.IMREAD_GRAYSCALE)
    perro_rojo = cv2.imread(perro_paths[2], cv2.IMREAD_GRAYSCALE)
    
    # Crear imágenes en color vacías
    imagen1_color = np.zeros((imagen1_azul.shape[0], imagen1_azul.shape[1], 3), dtype=np.uint8)
    imagen2_color = np.zeros((imagen2_azul.shape[0], imagen2_azul.shape[1], 3), dtype=np.uint8)
    perro_color = np.zeros((perro_azul.shape[0], perro_azul.shape[1], 3), dtype=np.uint8)
    
    # Asignar cada imagen de escala de grises a su canal de color
    imagen1_color[:, :, 0] = imagen1_azul  # Canal azul
    imagen1_color[:, :, 1] = imagen1_verde  # Canal verde
    imagen1_color[:, :, 2] = imagen1_rojo   # Canal rojo
    
    imagen2_color[:, :, 0] = imagen2_azul  # Canal azul
    imagen2_color[:, :, 1] = imagen2_verde  # Canal verde
    imagen2_color[:, :, 2] = imagen2_rojo   # Canal rojo
    
    perro_color[:, :, 0] = perro_azul  # Canal azul
    perro_color[:, :, 1] = perro_verde  # Canal verde
    perro_color[:, :, 2] = perro_rojo   # Canal rojo
    
    return imagen1_color, imagen2_color, perro_color

imagen1_paths = [
    'imagen1/imagen1_salida_gray_azul.jpg',
    'imagen1/imagen1_salida_gray_verde.jpg',
    'imagen1/imagen1_salida_gray_rojo.jpg'
]

imagen2_paths = [
    'imagen2/imagen2_salida_gray_azul.jpg',
    'imagen2/imagen2_salida_gray_verde.jpg',
    'imagen2/imagen2_salida_gray_rojo.jpg'
]

perro_paths = [
    'perro/perro_salida_gray_azul.jpg',
    'perro/perro_salida_gray_verde.jpg',
    'perro/perro_salida_gray_rojo.jpg'
]

# Reconstruyendo imagenes
imagen1_color, imagen2_color, perro_color = reconstruir_imagen_color(imagen1_paths, imagen2_paths, perro_paths)

fig, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].imshow(imagen1_color)
axs[0].set_title('Imagen 1 Reconstruida a Color')
axs[0].axis('off')

axs[1].imshow(imagen2_color)
axs[1].set_title('Imagen 2 Reconstruida a Color')
axs[1].axis('off')

axs[2].imshow(perro_color)
axs[2].set_title('Perro Reconstruida a Color')
axs[2].axis('off')

plt.show()
```

```python
def crear_escala_grises_3d(img3, output_prefix):
    
    imagen_color = cv2.imread(img3)
    
    azul, verde, rojo = cv2.split(imagen_color)
    
    cv2.imwrite(f'{output_prefix}_gray_azul.jpg', azul)
    cv2.imwrite(f'{output_prefix}_gray_verde.jpg', verde)
    cv2.imwrite(f'{output_prefix}_gray_rojo.jpg', rojo)
    
    return azul, verde, rojo

# Carga de la imagen de prueba
img3 = 'f16.tiff'
output_prefix = 'imagen_prueba'

# Creación de las escalas de grises
azul, verde, rojo = crear_escala_grises_3d(img3, output_prefix)

# Mostrar imagenes nuevas
fig, axs = plt.subplots(1, 4, figsize=(20, 5))

imagen_color = cv2.imread(img3)
imagen_color = cv2.cvtColor(imagen_color, cv2.COLOR_BGR2RGB) # Convertir de BGR a RGB para mostrar correctamente con matplotlib
axs[0].imshow(imagen_color)
axs[0].set_title('Imagen Original')
axs[0].axis('off')

axs[1].imshow(azul, cmap='gray')
axs[1].set_title('Canal Azul')
axs[1].axis('off')

axs[2].imshow(verde, cmap='gray')
axs[2].set_title('Canal Verde')
axs[2].axis('off')

axs[3].imshow(rojo, cmap='gray')
axs[3].set_title('Canal Rojo')
axs[3].axis('off')

plt.show()
```

```python
# Función del histograma de canal con medias aritméticas
img3 = 'f16.tiff'

def mostrar_histogramas_con_media(img3):
    # Leer la imagen en color
    imagen_color = cv2.imread(img3)
    
    # Separar los canales de color
    azul, verde, rojo = imagen_color[:, :, 0], imagen_color[:, :, 1], imagen_color[:, :, 2]
    
    # Convertir la imagen a escala de grises usando el promedio aritmético de los canales
    gris = (azul.astype(np.float32) + verde.astype(np.float32) + rojo.astype(np.float32)) / 3
    gris = gris.astype(np.uint8)
    
    # Función para calcular y mostrar histogramas sin usar OpenCV
    def mostrar_histograma(canal, color, titulo):
        hist, bins = np.histogram(canal.flatten(), bins=256, range=[0, 256])
        media = np.mean(canal)
        
        plt.plot(hist, color=color)
        plt.axvline(x=media, color='k', linestyle='dashed', linewidth=1)
        plt.title(titulo)
        plt.xlim([0, 256])
        plt.xlabel('Intensidad')
        plt.ylabel('Frecuencia')

    # Mostrar los histogramas
    plt.figure(figsize=(20, 10))

    # Histograma del canal azul
    plt.subplot(2, 2, 1)
    mostrar_histograma(azul, 'blue', 'Histograma del Canal Azul')

    # Histograma del canal verde
    plt.subplot(2, 2, 2)
    mostrar_histograma(verde, 'green', 'Histograma del Canal Verde')

    # Histograma del canal rojo
    plt.subplot(2, 2, 3)
    mostrar_histograma(rojo, 'red', 'Histograma del Canal Rojo')

    # Histograma de la imagen en escala de grises
    plt.subplot(2, 2, 4)
    mostrar_histograma(gris, 'gray', 'Histograma de la Imagen en Escala de Grises')

    plt.tight_layout()
    plt.show()

mostrar_histogramas_con_media(img3)
```

## Problema 5

Investigue en que consiste el enfoque de escala de grises ponderado, luego de esto implemente una funcion que dada una imagen, realice una escala de grises ponderada (notar que no existe una

### Enfoque de escala de grises ponderado.
Este enfoque consiste en tomar una imagen a color y convertirla a escala de grises, utilizando una combinación ponderada de los canales rojo, verde y azul. Esta ponderación se hace tomando en cuenta la precepción humana de los colores, ya que no todos contribuyen de igual manera a la percepción del brillo.

#### Coeficientes Ponderados Comunes
Los coeficientes comúnmente usados para la conversión ponderada son:

- Rojo: 0.299
- Verde: 0.587
- Azul: 0.114

En este orden, el verde es el que mayor contribución tiene a la percepción del brillo, seguido del rojo y de último el azul.

```python
def convertir_a_gris_ponderado(img3):
   
    imagen_color = cv2.imread(img3)
    
    # Separar los canales de color
    azul, verde, rojo = imagen_color[:, :, 0], imagen_color[:, :, 1], imagen_color[:, :, 2]
    
    # Coeficientes de ponderación
    coef_azul = 0.114
    coef_verde = 0.587
    coef_rojo = 0.299
    
    # Convertir la imagen a escala de grises ponderada
    gris_ponderado = (coef_azul * azul + coef_verde * verde + coef_rojo * rojo).astype(np.uint8)
    
    return gris_ponderado

gris_ponderado = convertir_a_gris_ponderado(img3)

# Mostrar imagenes
imagen_color = cv2.imread(img3)
imagen_color = cv2.cvtColor(imagen_color, cv2.COLOR_BGR2RGB) 

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(imagen_color)
plt.title('Imagen Original')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(gris_ponderado, cmap='gray')
plt.title('Imagen en Escala de Grises Ponderada')
plt.axis('off')

plt.tight_layout()
plt.show()
```

### Espacio de color HSV

Es un modelo de color que describe colores en tres componentes: Hue (matiz), Saturación y Valor (o brillo). 

Es especialmente util en aplicaciones de procesamiento de imagenes porque este modelo toma la forma en la que los humanos perciben y describen los colores.

1. Hue (matiz): Representa el tipo de color. Se mide en grados del círculo cromático, que van del 0 al 360, siendo 0° el rojo y 240° el azul.
2. Saturación: Indica la intensidad o pureza del color. Va de 0 a 100% siendo 0 un tono gris sin matiz y el 100% el color puro.
3. Valor o brillo: Representa el brillo y va de 0 a 100%, siendo 0 completamente negro y 100 el valor mas brillante posible.

PAra mapear los colores al espacio HSV se debe seguir el proceso siguiente:
1. Normalizar los valores RGB
2. Calcular el valor V
3. Calcular la saturación S
4. Calcular el matiz H
